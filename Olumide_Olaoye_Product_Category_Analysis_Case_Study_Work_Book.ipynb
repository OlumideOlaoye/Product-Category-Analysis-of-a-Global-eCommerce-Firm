### Question 1: What was the most profitable month for the company? (Assume constant commission rate across different categories.)

```python
## To get the most profitable month, we check the relationship between the month and net_transaction_value(euros)
plt.figure(figsize=(15,10))
sns.barplot(x = 'Month', y = 'net_transaction_value(euros)', data = category_data)
```

### Question 2: What were the top 3 selling categories for the last quarter of 2015? How does it compare to the first quarter of the same year?

### Question 9: Which category has the best net/gross ratio. What do you think this means?


```python
category_data.head()
```

```python
## Create a new column 'net/gross ratio' which is 'net_trasaction_value(euros) / gross_transaction_value(euros)'
category_data['net/gross ratio'] = category_data['net_transaction_value(euros)'] / category_data['gross_transaction_value(euros)']
```

```python
category_data.head(5)
```

```python
## To get the category with the best net/gross ratio
plt.figure(figsize=(17,15))
sns.barplot(x='net/gross ratio', y='Product_category', data=category_data)
```

**Jewelry Cleaning & Care** and **Gadgets** have the highest net/gross ratio among the product categories. 

What this means is that the cost incurred on the sales and delivery of Jewelry Cleaning & Care and Gadgets are small compared to other product categories.
